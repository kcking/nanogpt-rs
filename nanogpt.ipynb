```python
import nanogpt
import matplotlib.pyplot as plt

nanogpt.text

def plot_char_counts():
    char_counts = {}
    for char in nanogpt.text:
        if char in char_counts:
            char_counts[char] += 1
        else:
            char_counts[char] = 1

    char_counts = dict(sorted(char_counts.items(), key=lambda item: item[1], reverse=True))
    plt.figure(figsize=(20,5))
    plt.bar(range(len(char_counts)), list(char_counts.values()), align='center')
    plt.xticks(range(len(char_counts)), [k.replace('\n', '\\n').replace(' ', '\' \'') for k in char_counts.keys()])

plot_char_counts()
```

```python
import nanogpt
import nanogpt_rs
import timeit

# Test correctness
assert(set(nanogpt.chars) == nanogpt_rs.unique_characters(nanogpt.text))

iters = 100
py_time = timeit.timeit('set(nanogpt.text)', setup="import nanogpt", number=iters)
rs_time = timeit.timeit('nanogpt_rs.unique_characters(nanogpt.text)', setup="import nanogpt, nanogpt_rs", number=iters)

print(f"python: {py_time/iters:.4f}s/iter, rust: {rs_time/iters:.4f}s/iter")
```
